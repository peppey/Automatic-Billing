```python
##### Für Erinnerung
import pandas as pd
data = pd.read_csv('test.asc', encoding='cp850')
from datetime import date
```

```python
# Leere Einträge auf "" setzen
for i in range(len(data.columns)):
    if "Unnamed" in data.columns[i]:
        data.columns.values[i] = ""  
        
data.columns
```

```python
# Unser Zeichen zusammenfügen
number = data.columns[13]+" - "+data.columns[14]+" - "+data.columns[15]
```

```python
from PyPDF2 import PdfFileWriter, PdfFileReader
import io
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import letter


# Ansprechpartner oben links #Mrsr Patentenheten
# Straße oben links
# Patentnummer leer?

# P 600 32 450.4
# Schweden UND Deutschland?
# 0, 0.2, 0.3, 0.4
# P 06-257:3
# 2151




packet = io.BytesIO()
# create a new PDF with Reportlab
can = canvas.Canvas(packet, pagesize=letter)
can.drawString(300, 427, data.columns[5]) # Jahr
# Fälligkeit, nur Tag+Datum extrahieren
splittedDate = data.columns[25].split(".")
dayAndMonth = splittedDate[:2]
joinedDate = ".".join(dayAndMonth)+"."
can.drawString(300, 450, joinedDate) # Fälligkeit
can.drawString(300, 462, data.columns[24]) # Anmeldetag
can.drawString(300, 474, data.columns[23]) # Bezeichnung, 1. Zeile
can.drawString(300, 485, data.columns[22]) # Bezeichnung, 2. Zeile
can.drawString(300, 569, number) # Unser Zeichen
can.drawString(300, 557, data.columns[16]) # Inhaber
can.drawString(300, 521, data.columns[19]) # Land
can.drawString(300, 509, data.columns[20]) # Anmeldenummer
can.drawString(300, 497, data.columns[21]) #Patentnummer
can.drawString(300, 545, data.columns[17]) #Sitz
can.drawString(300, 533, data.columns[18]) #Schutzrecht


# Oben links
can.drawString(90, 680, data.columns[16]) #Inhaber
name = data.columns[27] + " " + data.columns[29] #Name zusammenfügen
can.drawString(90, 668, name) #Anrede
can.drawString(90, 656, data.columns[30]) #Straße 1
can.drawString(90, 644, data.columns[31]) #Straße 2
can.drawString(90, 632, data.columns[32]) #Ort
can.drawString(90, 620, data.columns[33]) #Land


### Unten links
can.drawString(60, 220, data.columns[13]) # Unser Zeichen, 1. Teil
can.drawString(90, 220, "-") #Unser Zeichen, 1. Strich
can.drawString(95, 220, data.columns[14]) # Unser Zeichen, 2. Teil
can.drawString(100, 220, "-") #Unser Zeichen, 2. Strich
can.drawString(105, 220, data.columns[15]) # Unser Zeichen, 3. Teil
can.drawString(60, 208, data.columns[16]) #Inhaber
can.drawString(60, 196, data.columns[17]) #Sitz
can.drawString(60, 184, data.columns[18]) #Schutzrecht
can.drawString(60, 172, data.columns[19]) #Land
can.drawString(60, 160, data.columns[20]) #Anmeldenummer
can.drawString(60, 136, data.columns[22]) #Bezeichnung, 2. Zeile #evtl existieren nicht überall zwei Zeilen, testen
can.drawString(60, 124, data.columns[23]) #Bezeichnung, 1. Zeile
can.drawString(60, 112, data.columns[24]) #Anmeldetag
can.drawString(60, 100, data.columns[25]) #Fälligkeit
can.drawString(60, 88, data.columns[5]) #Jahr


#Kosten
can.drawString(90, 470, data.columns[0])

#Heutiges Datum
# in deutsches Format umwandeln
splittedDateToday = str(date.today()).split("-")
splittedDateToday.reverse()
joinedDateToday = ".".join(splittedDateToday)
can.drawString(390, 700, joinedDateToday)


can.save()

#move to the beginning of the StringIO buffer
packet.seek(0)
new_pdf = PdfFileReader(packet)
# read your existing PDF
existing_pdf = PdfFileReader(open("Erinnerung.pdf", "rb"))
output = PdfFileWriter()
# add the "watermark" (which is the new pdf) on the existing page
page = existing_pdf.getPage(0)
page.mergePage(new_pdf.getPage(0))
output.addPage(page)
# finally, write "output" to a real file
outputStream = open("Erinnung_neu.pdf", "wb")
output.write(outputStream)
outputStream.close()
```
