```python
import pandas as pd
data = pd.read_csv('PIA.asc', encoding='cp850')
from datetime import date
```

```python
# Leere Einträge auf "" setzen
for i in range(len(data.columns)):
    if "Unnamed" in data.columns[i]:
        data.columns.values[i] = ""  
        
data.columns
```

```python
from PyPDF2 import PdfFileWriter, PdfFileReader
import io
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import letter

packet = io.BytesIO()


### Fragen:
# Nr. unten links
# Betrag EUR 16 Prozent Mehrwertsteuer


canFour = canvas.Canvas(packet, pagesize=letter)

# Unten links
canFour.drawString(80, 311, data.columns[16]) # Auftraggeber
canFour.drawString(165, 311, data.columns[29]) # Ansprechpartner
canFour.drawString(175, 285, data.columns[16]) # Auftraggeber
number = data.columns[13]+" - "+data.columns[14]+" - "+data.columns[15] #Zeichen Auftraggeber
canFour.drawString(175, 297, number) # Zeichen Auftraggeber
canFour.drawString(175, 274, data.columns[17]) # Sitz
canFour.drawString(175, 262, data.columns[18]) # Schutzrecht
canFour.drawString(175, 250, data.columns[19]) # Land
canFour.drawString(175, 238, data.columns[20]) # leer
canFour.drawString(175, 226, data.columns[21]) # Patent-Nr
canFour.drawString(175, 214, data.columns[22]) # Titel
canFour.drawString(175, 202, data.columns[23]) # leer
canFour.drawString(175, 190, data.columns[24]) # Anmeldetag
# Fälligkeit, nur Tag+Datum extrahieren
splittedDate = data.columns[25].split(".")
dayAndMonth = splittedDate[:2]
joinedDate = ".".join(dayAndMonth)+"."
canFour.drawString(175, 178, joinedDate) # Fälligkeit


# Oben links
canFour.drawString(85, 700, data.columns[27]) # "Firma"
canFour.drawString(85, 688, data.columns[16]) # Auftraggeber
canFour.drawString(85, 676, data.columns[29]) # Ansprechpartner
canFour.drawString(85, 664, data.columns[30]) # leer
canFour.drawString(85, 652, data.columns[31]) # Straße
canFour.drawString(85, 640, data.columns[17]) # Sitz
canFour.drawString(85, 628, data.columns[33]) # Land

# Oben rechts
#Heutiges Datum
# in deutsches Format umwandeln
splittedDateToday = str(date.today()).split("-")
splittedDateToday.reverse()
joinedDateToday = ".".join(splittedDateToday)
canFour.drawString(390, 715, joinedDateToday)

# Unten rechts
canFour.drawString(390, 320, joinedDateToday)




canFour.save()

#move to the beginning of the StringIO buffer
packet.seek(0)
new_pdf = PdfFileReader(packet)
# read your existing PDF
existing_pdf = PdfFileReader(open("Rechn.pdf", "rb"))
output = PdfFileWriter()
# add the "watermark" (which is the new pdf) on the existing page
page = existing_pdf.getPage(0)
page.mergePage(new_pdf.getPage(0))
output.addPage(page)
# finally, write "output" to a real file
outputStream = open("Rechn_neu.pdf", "wb")
output.write(outputStream)
outputStream.close()
```
